```python
import hyperbolicTSNE.uniform_grid as grid
import numpy as np
from matplotlib import pyplot as plt

euclidean_distances = np.linspace(0.0, 0.999, 500)
hyperbolic_distances = [grid.distance_py(e_dist, 0.0, 0.0, 0.0) for e_dist in euclidean_distances]

# Plotting the distances
plt.plot(euclidean_distances, hyperbolic_distances)
plt.title('Euclidean vs. Hyperbolic Distances')
plt.xlabel('Euclidean Distance')
plt.ylabel('Hyperbolic Distance')
plt.grid(True)
plt.show()
```
