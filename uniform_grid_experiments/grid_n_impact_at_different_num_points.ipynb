```python
from helpers import run, run_mult, Datasets, plot_points, plot_points_separate
import numpy as np

data_home = "../datasets"
log_path = "temp/poincaregpu/"

only_animate = False
seed = 42
perplexity = 30
logging = False

dataset = Datasets.MNIST

# TEST FOR DIFFERENT GRID_N VALUES
u_grid_ns = [10, 20, 40, 60, 80, 100, 120, 140, 200, 300, 500]
num_points_arr = [1000, 4000, 10000, 30000, 70000]

ugrid_gpu_precisions = []
ugrid_gpu_recalls = []
ugrid_gpu_runtimes = []
```

```python

for num_points in num_points_arr:
    precisions = []
    recalls = []
    runtimes = []
    for i, u_grid_n in enumerate(u_grid_ns):
        p, r, t = run_mult(3, num_points, False, True, u_grid_n, logging, log_path, dataset, perplexity, data_home) # Running approx. using uniform grid
        precisions.append(np.mean(p))
        recalls.append(np.mean(r))
        runtimes.append(t)

    ugrid_gpu_precisions.append(precisions)
    ugrid_gpu_recalls.append(recalls)
    ugrid_gpu_runtimes.append(runtimes)

```
